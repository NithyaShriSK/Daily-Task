Part 5 – Calculus

1.Assume:
 Wellbeing_Index = (GDP×Happiness) + 0.5×(Happiness−5)^2
 Find derivative wrt Happiness.

```python
import sympy as sp
gdp, ladder = sp.symbols('gdp ladder')
wellbeing_index = (gdp * ladder) + (0.5 + ladder**2)
derivative_expr = sp.diff(wellbeing_index, ladder)
print(derivative_expr)
```

Part 6 – Feature Engineering

1.Create is_happy = 1 if happiness score > 7 else 0.
2.Create happiness_rank based on descending Ladder score.
3.Create above_average_gdp = 1 if GDP > global mean.
4.Create quartiles based on Healthy life expectancy.

```python
df["is_happy"]=df["Ladder score"].apply(lambda x:1 if x>7 else 0)
df.head()
```

```python
df["happiness_rank"]=df["Ladder score"].rank(ascending=False).astype(int)
df.head()
```

```python
mean=df["Log GDP per capita"].mean()
df["above_average_gdp"]=df["Log GDP per capita"].apply(lambda x: 1 if x>mean else 0)
df.head()
```

```python
df["Quartiles"]=pd.qcut(df["Healthy life expectancy"],q=4)
df.head()
```

```python
meanresidual=df["Dystopia + residual"].mean().astype(float)
df["Dystopia + residual"].fillna(meanresidual,inplace=True)
```

```python
meanquart=df["Quartiles"].mode()
df["Quartiles"].fillna(meanquart,inplace=True)
```

```python
plt.boxplot(df["Ladder score"])
plt.show()
```

```python
plt.boxplot(df["Log GDP per capita"])
plt.show()
```

```python
q1=df["Ladder score"].quantile(0.25)
q3=df["Ladder score"].quantile(0.75)
IQR=q3-q1
lower=q1-(1.5*IQR)
higher=q3+(1.5*IQR)
mean=df["Ladder score"].mean()
outliers=df[(df["Ladder score"]<lower)|(df["Ladder score"]>higher)]["Ladder score"]
df["Ladder score"]=df["Ladder score"].apply(lambda x: mean if (x<lower or x>higher) else x)
```

```python
plt.boxplot(df["Ladder score"])
plt.show()
```

```python
q1=df["Log GDP per capita"].quantile(0.25)
q3=df["Log GDP per capita"].quantile(0.75)
IQR=q3-q1
lower=q1-(1.5*IQR)
higher=q3+(1.5*IQR)
mean=df["Log GDP per capita"].mean()
outliers=df[(df["Log GDP per capita"]<lower)|(df["Log GDP per capita"]>higher)]["Log GDP per capita"]
df["Log GDP per capita"]=df["Log GDP per capita"].apply(lambda x: mean if (x<lower or x>higher) else x)
```
